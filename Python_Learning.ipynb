```python
# Assigned by reference
# Creating a list
original_list = [1, 2, 3]

# Assigning it to another variable
copied_list = original_list

print(original_list)  # Output: [1, 2, 3]
print(copied_list)    # Output: [1, 2, 3]
# Modifying the copied list
copied_list.append(4)

print(original_list)  # Output: [1, 2, 3, 4]
print(copied_list)    # Output: [1, 2, 3, 4]
```

```python
# Using slicing
copied_list = original_list[:]

# Using copy method
copied_list = original_list.copy()
copied_list.append(4)
print(original_list)  # Output: [1, 2, 3]
print(copied_list)    # Output: [1, 2, 3, 4]
```

# EXCEPTION HANDLING

try: (add code which we to run and error can occur)

except Exception as ex: (it will print error that might occur)

else: (this will only print if no exception occur)

finally: (this will print at the end of the code)

1. Errors

- Detection: Occur before the program starts running, during compilation or interpretation.
- Types:
- Syntax errors: Incorrect grammar or structure of the code.
- Logical errors: Incorrect logic in the code, leading to unexpected results.
- Handling: Usually require fixing the code itself to resolve.

2. Exceptions
- Detection: Occur during program execution, when something unexpected happens.
- Types:
- Runtime errors: Issues that arise during the execution of the program.
- Specific exceptions: Python provides various built-in exception classes (e.g., ZeroDivisionError, TypeError, ValueError) for different error conditions.
- Handling: Can be handled using try-except blocks to gracefully recover from the error or provide alternative actions.

```python
try:
    ##code block where exception can occur
    a=int(input("Enter the number 1 "))
    b=int(input("Enter the number 2 "))
    c=a/b

except NameError:
    print("The user have not defined the variable")
except ZeroDivisionError:
    print("Please provide number greater than 0")
except TypeError:
    print("Try to make the datatype similar")
except Exception as ex:
    print(ex)
else:
    print(c)
finally:
    print("The execution is done")
```

# ASSERT STATEMENT
Python provides the assert statement to check if a given logical expression is true or false. Program execution proceeds only if the expression is true and raises the AssertionError when it is false. The following code shows the usage of the assert statement.

```python
num=10
assert num>10
```

```python
try:
    num=int(input("Enter the even number"))
    assert num%2==0
    print("The number is even")
except AssertionError:
    print("Please enter even number")
```
