```python
df_country_name = df.Country.value_counts().index
```

```python
df_country_value = df.Country.value_counts().values
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(df_country_value[:3], labels = df_country_name[:3], autopct = '%1.2f%%')
```

```python
df.columns
```

```python
ratings = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

# Observation
1. When Rating is between 4.5 to 4.9---> Excellent
2. When Rating are between 4.4 to 3.4--->very good
3. when Rating is between 3.5 to 3.9----> good
4. when Rating is between 3.0 to 3.4----> average
5. when Rating is between 2.5 to 2.9----> average
6. when Rating is between 2.0 to 2.4----> Poor

```python
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings, hue = 'Rating color')
```

Observation:

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4
